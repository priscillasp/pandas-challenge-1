## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df["line_subtotal"] = df["unit_price"]* df["qty"]
df.loc[ 0:1 , ["unit_price", "qty", "line_subtotal"]]
```

```python
#Created a total weight column 
df["total_weight"] = df["unit_weight"]* df["qty"]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# CPP = Cost Per Pound 

def cpp (total_weight): 
    if total_weight > 50:
        return total_weight * 7
    else:
        return total_weight * 10
                

df["shipping_price"] = df["total_weight"].apply(cpp)

df.loc[0:2 , ["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]]
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

sum_st_sp = df["line_subtotal"] + df["shipping_price"]
calculate_tax = sum_st_sp * 0.0925
df["line_price"] = (sum_st_sp + calculate_tax).round(2)

df.loc[0:2 , ["line_subtotal", "shipping_price", "line_price"]]
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df["line_cost"] = (df["unit_cost"] * df["qty"]) + df["shipping_price"]
df.loc[0:2 , ["unit_cost","qty", "shipping_price", "line_cost"]]
```

```python
# Create a column for the profit of
# each line using line cost and line price

df["line_profit"] = df["line_price"] - df["line_cost"]
df.loc[0:2 , ["line_price","line_cost", "line_profit"]]
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_5_complete_df = df[df["client_id"].isin(top_5_list)]


top_spenders_df = top_5_complete_df.groupby("client_id")["line_price"].sum()


print(top_spenders_df)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

top_spenders_summary_df = top_5_complete_df.groupby("client_id", as_index=False).agg(
    qty=('qty', 'sum'),  
    shipping_price=('shipping_price', 'sum'),  
    line_price=('line_price', 'sum'), 
    line_cost=('line_cost', 'sum'),
    line_profit=('line_profit', 'sum')  
    ).sort_values('line_profit', ascending=False)
top_spenders_summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.

top_spenders_summary_df = top_spenders_summary_df.rename(columns = {"client_id":"Client ID",
                                                                 "qty":"Units",
                                                                 "shipping_price":"Shipping",
                                                                 "line_price": "Total Revenue",
                                                                 "line_cost": "Total Cost",
                                                                 "line_profit":"Total Profit"})
top_spenders_summary_df
```

```python
# Currency should be in millions of dollars.
def format_currency_millions(x):
    return "${:,.2f}M".format(x / 1_000_000)
formatted_columns_df = top_spenders_summary_df[['Shipping', 'Total Revenue', 'Total Cost', 'Total Profit']].apply(lambda x: x.apply(format_currency_millions))
formatted_columns_df
```

```python
#concat data with newly formatted currency
top_spenders_summary_df = top_spenders_summary_df[["Client ID", "Units"]]
top_spenders_final_df = pd.concat([top_spenders_summary_df, formatted_columns_df], axis=1)
top_spenders_final_df
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
top_spenders_final_df.sort_values(["Total Profit"], ascending=False)
```
